```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

import warnings
warnings.filterwarnings("ignore")
```

```python
df=pd.read_csv("Students.csv")
df
```

```python
df["Total"]=df["Python"]+df["Machine Learning"]
df
#created a column Total
```

```python
cities=["Mumbai","Mumbai","Mumbai","Banglore","Delhi","Delhi","Delhi","Mumbai","Delhi","Goa"]
df["City"]=cities
df
#aded a column of cities
```

```python
def getgrade(marks):
    v=marks/2
    if v>=75:
        return "A"
    elif v>=60 and v<75:
        return "B"
    elif v>=35 and v<60:
        return "C"
    else:
        return "F"

df["Grade"]=df["Total"].apply(getgrade)
df
#grade column has been added based on total marks of students.
```

```python
# what is the avg score of students who live in mumbai
df[df["City"]=="Mumbai"]["Total"].mean()
```

```python
#what is the avg score of students who live in delhi
df[df["City"]=="Delhi"]["Total"].mean()
```

```python
df.groupby("City")["Total"].mean().sort_values(ascending=True).plot(kind="bar")
#Stuents in city of Mumbai are having the maximum score around 150
#Students in city of Goa have min score of 90
```

```python
df.sort_values(["Total"]).agg(["mean","min","max","count"])
#Minimum score is of the student chandani,she has non-tech background,her city is bangalore and total score is 91.0
#Maximum score is of the student Shani,He has a tech background,his city is Mumbai and total score 184.0
```

```python
mp=df["Total"].mean()
mep=df["Total"].median()
sns.distplot(df["Total"])
plt.axvline(mp,color='r')
plt.axvline(mep,color='y')
plt.grid(True)
#mean=red line
#median=yellow line
#difference between median and mean is less.
#but mean>median=Positive skewness.
```
